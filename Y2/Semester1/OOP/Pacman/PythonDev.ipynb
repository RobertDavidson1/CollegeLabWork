```python
class Assets:
    # Define ANSI escape codes for True Color using RGB values
    RESET = "\033[0m"

    # Ghosts' colors (RGB)
    PINK = "\033[38;2;234;130;229m"  # Pink
    LIGHT_BLUE = "\033[38;2;70;191;238m"  # Light blue
    RED = "\033[38;2;255;62;25m"  # Red
    ORANGE = "\033[38;2;219;133;28m"  # Orange

    # Assets and Walls' colors (RGB)
    YELLOW = "\033[38;2;253;255;0m"  # Yellow
    BLUE = "\033[38;2;33;33;222m"  # Blue

    PACMAN = 'P'
    GHOST = 'G'
    WALL = '▀'
    FOOD = '•'
    EMPTY = ' '
```

```python

def show_grid(grid):
    food_remaining = 0
    for row in grid:
        coloured_row = []
        for cell in row:
            if cell == Assets.PACMAN:
                coloured_row.append(f"{Assets.YELLOW}{cell}{Assets.RESET}")
            elif cell == Assets.GHOST:
                coloured_row.append(f"{Assets.RED}{cell}{Assets.RESET}")
            elif cell == Assets.WALL:
                coloured_row.append(f"{Assets.BLUE}{cell}{Assets.RESET}")
                food_remaining += 1
            elif cell == Assets.FOOD:
                coloured_row.append(f"{Assets.RESET}{cell}{Assets.RESET}")

            else:
                coloured_row.append(cell)
        print(' '.join(coloured_row))
    return food_remaining
    
```

```python
def init_grid(grid_height=9, grid_width = 30):

    HEIGHT = min(max(9,grid_height), 15) 
    WIDTH = min(max(grid_width, 30), 35)

    #  Ensure odd height and width for ideal centering
    if HEIGHT % 2 != 1:
        HEIGHT += 1
    if WIDTH % 2 != 1:
        WIDTH +=1


    grid = [[Assets.FOOD for _ in range(WIDTH)] for _ in range(HEIGHT)]

    for i in range(WIDTH):
        grid[0][i] = Assets.WALL
        grid[HEIGHT - 1][i] = Assets.WALL


    for i in range(HEIGHT):
        grid[i][0] = Assets.WALL
        grid[i][WIDTH - 1] = Assets.WALL

    # 5 x 23 art
    art = [
        ['▀','▀','▀','•','▀','▀','▀','•','▀','▀','•','•','▀','▀','•','•','•','▀','•','•','▀','▀',"•"],
        ['▀','•','•','•','▀','•','•','•','•','•','▀','•','•','▀','•','•','▀',' ','▀','•','•','▀',"•"],
        ['▀','•','•','•','▀','▀','▀','•','•','•','▀','•','•','▀','•','•','▀',' ','▀','•','•','▀',"•"],
        ['▀','•','•','•','•','•','▀','•','▀','•','•','•','•','▀','•','•','▀',' ','▀','•','•','▀',"•"],
        ['▀','▀','▀','•','▀','▀','▀','•','▀','▀','▀','•','▀','▀','▀','•','•','▀','•','•','▀','▀','▀'],
    ]

    
    start_x = (HEIGHT - 5) // 2 # starting x for the art
    start_y = (WIDTH - 23) // 2 # starting y for the art

    for i in range(len(art)):  # Loop over rows of art
        for j in range(len(art[i])):  # Loop over columns of each row
            grid[start_x + i][start_y + j] = art[i][j]

    grid[HEIGHT // 2][1] = Assets.PACMAN
    grid[HEIGHT // 2][-2] = Assets.GHOST

    return grid
```
